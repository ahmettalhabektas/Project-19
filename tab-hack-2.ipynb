```python
tr=pd.get_dummies(train)
```

```python
te=pd.get_dummies(test)
```

```python
tr_cols=tr.columns
te_cols=te.columns
```

```python
list=[]
for i in tr_cols:
    if not i in te_cols:
        list.append(i)
list
```

```python
y_train.value_counts()
```

```python
g=GaussianNB()
b=BernoulliNB()
k=KNeighborsClassifier(n_neighbors=10)
d=DecisionTreeClassifier()
gbc=GradientBoostingClassifier()
```

```python
algorithms = [g,b,k,d,gbc]
```
